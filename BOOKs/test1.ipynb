```python
from pydantic import BaseModel, Field
from typing import Optional
import datetime

class book(BaseModel):
    id : Optional[int] = None
    name : str = Field(min_length=1)
    rating : int = Field(ge=1,le=5)
    date : int= Field(ge = 1900, le = datetime.date.today().year)

books: list[book] = []

def find_book_id(Book: book) -> book:
    if(not len(books)):
        book.id = 1
    else:
        book.id = books[-1].id +1
    return book

def add_book(Book:book):
    books.append(find_book_id(book))
```

```python
import asyncio
import aiohttp

def fetch_url(session, url):
    async with session.get(url) as response:
        data = await response.json()
        return data

async def fetch_all(urls):
    async with aiohttp.ClientSession() as session:
        tasks = [fetch_url(session, url) for url in urls]
        results = await asyncio.gather(*tasks)
        return results

async def main():
    urls = [
        "https://jsonplaceholder.typicode.com/posts/1",
        "https://jsonplaceholder.typicode.com/posts/2",
        "https://jsonplaceholder.typicode.com/posts/3",
        "https://jsonplaceholder.typicode.com/users/1",
        "https://jsonplaceholder.typicode.com/todos/1",
    ]

    print(f"Fetching {len(urls)} URLs concurrently...\n")

    import time
    start = time.time()
    results = await fetch_all(urls)
    elapsed = time.time() - start

    for i, data in enumerate(results):
        print(f"--- Result {i+1} ---")
        print(f"  URL: {urls[i]}")
        print(f"  Keys: {list(data.keys())}")
        print(f"  Title/Name: {data.get('title') or data.get('name') or data.get('todo')}")
        print()

    print(f"✅ All {len(urls)} requests done in {elapsed:.2f}s")

result = await main()
print(result)
```
